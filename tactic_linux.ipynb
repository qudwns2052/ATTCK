```python
import pandas as pd
import numpy as np
import json

pd.set_option('display.max_row', 500)
pd.set_option('display.max_columns', 100)
```

```python
f=open("result.json",'r')
jsondata=json.load(f)
f.close()
```

```python
df=pd.DataFrame(jsondata,index=['count'])
df=df.T
df.head()
```

```python
import os
from selenium import webdriver
from bs4 import BeautifulSoup

path = os.getcwd() + "/chromedriver"
driver = webdriver.Chrome(path)

url = "https://attack.mitre.org/tactics/enterprise/"
driver.get(url)
driver.implicitly_wait(10)

html = driver.page_source  
bs = BeautifulSoup(html, "html.parser")
table = bs.find("table", class_ = "table").find("tbody")
repos = table.find_all("tr")

tactics = []
for repo in repos :
    line = repo.find_all("a")
    tactics.append([line[0].text, line[1].text])

result = []

techs_max = 0
for tactic in tactics :
    url = "https://attack.mitre.org/tactics/" + tactic[0] + "/"
    driver.get(url)
    driver.implicitly_wait(10)
    html = driver.page_source  
    bs = BeautifulSoup(html, "html.parser")
    techs = bs.find("table", class_ = "table").find("tbody").find_all("tr")
    techs_id = []
    for tech in techs :
        tech_id = tech.find("a").text.replace(' ', '')
        if jsondata.get(tech_id) :
            techs_id.append(tech_id + ":" + str(jsondata[tech_id]))
    
    if (len(techs) > techs_max) :
        techs_max = len(techs)
    result.append([tactic, techs_id])

driver.quit()

df_result=pd.DataFrame()

for tactic_info in result :
    tactic_info[1] = sorted(tactic_info[1], key=lambda tactic_info: int(tactic_info.split(":")[-1]), reverse=True)
    while len(tactic_info[1]) < techs_max :
        tactic_info[1].append(" ")
    df_result[tactic_info[0][1]] = tactic_info[1]


df_result
```
